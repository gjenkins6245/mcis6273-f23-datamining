**&#167; Task:**  (_Frequent Pattern Mining_) 

Which _early evening_ species are frequent together in groups of 2 or more regardless
of rural or urban? (Use a support of $0.50$ when answering).
  
**Response:** See code below. The following sets of species are frequently together:
 - Domestic Cat, Raccoon
 - Domestic Cat, White-tailed Deer
 - Raccoon, Nine-banded Armadillo
 - White-tailed Deer, Nine-banded Armadillo
 - White-tailed Deer, Raccoon
 - Domestic Cat, White-tailed Deer, Raccoon
 - White-tailed Deer, Raccoon, Nine-banded Armadillo

````python
import pandas as pd
from mlxtend.frequent_patterns import apriori

# Read in the cleaned dataset
freq1  = pd.read_csv('original_dataset_cleaned.csv')

# Filter 'Time of Detection' to get times between 2100 (9p) and 2359 (1159p)
freq1 = freq1[freq1['Time of Detection'] >= 210000]

# Group by 'Date of Detection' & 'Species Detected' and sum 'Number of the Species Detected in the Photo'.
freq1 = freq1.groupby(['Date of Detection','Species Detected']).sum()[['Number of the Species Detected in the Photo']]

# Change the summed values greater than 0 to True
freq1 = freq1.map(lambda x: True if x > 0 else False)

# Unstacking data to convert species to column headings and sorting by date
freq1 = freq1.unstack(sort='Date of Detection')

# Replacing nan data with False
freq1 = freq1.fillna(False)

# Removing extra column heading 'Number of the Species Detected in the Photo'
freq1.columns = freq1.columns.droplevel()

# Complete pattern mining
freq1_is = apriori(freq1, min_support=0.5, use_colnames=True)

# Calculate itemset length
freq1_is['length'] = freq1_is['itemsets'].apply(lambda x: len(x))

# Filter itemset length >= 2
freq1_is = freq1_is[freq1_is['length'] > 1]

# Display dataframe
freq1_is
````
